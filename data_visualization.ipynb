```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# .linspace() function returns evenly spaced numbers over a specified interval.
x = np.linspace(0, 20, 100)
```

```python
print(x)
```

```python
plt.figure(figsize = (9, 6))
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'b--')

plt.legend(('sin', 'cos'))
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), 'y-')
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), 'r--', linewidth = 3)

plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(13,13))
fig.tight_layout(pad=8)
fig.suptitle("Histogram Plots")
# object veri tipi hariç veri tiplerine sahip olan kolonları liste olarak tut
cols = iris.select_dtypes(exclude = 'object').columns

x_axes=0
y_axes=0

for col in cols:
  sns.histplot(data = iris, x = col, kde = True, ax = axes[x_axes, y_axes])
  if y_axes == 1:
    y_axes=0
    x_axes+=1
  else:
    y_axes +=1

plt.show()
```
