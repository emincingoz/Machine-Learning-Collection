# PCA (Principal Component Analysis)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/emincingoz/Machine-Learning-Collection/main/Datasets/Wine.csv')
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
x = dataset.iloc[:, 0:13].values
y = dataset.iloc[:, 13].values     # Customer_Segment
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)

print(x_train_scaled.shape)
```

```python
from sklearn.decomposition import PCA

# n_components: Selected how many dimensions the data will be reduced to
# Same PCA model used to transform x_train and x_test
pca = PCA(n_components = 2)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)

print(x_train_pca.shape)
```

```python
print("Variance Ratio: ", pca.explained_variance_ratio_)
print("Sum: ", sum(pca.explained_variance_ratio_))
# By choosing 2 components for the transformation, approximately 56% of the total variance is preserved
```

```python
from sklearn.linear_model import LogisticRegression

# Since it will be used twice, random_state = 0 was determined to test on the same model
lr_before_pca = LogisticRegression(random_state = 0)
lr_before_pca.fit(x_train_scaled, y_train)

lr_after_pca = LogisticRegression(random_state = 0)
lr_after_pca.fit(x_train_pca, y_train)
```

```python
y_pred_before_pca = lr_before_pca.predict(x_test_scaled)

y_pred_after_pca = lr_after_pca.predict(x_test_pca)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm_before_pca = confusion_matrix(y_test, y_pred_before_pca)
sns.heatmap(cm_before_pca, cmap = 'Blues', annot = True)
plt.title('Before PCA')
plt.show()

cm_after_pca = confusion_matrix(y_test, y_pred_after_pca)
sns.heatmap(cm_after_pca, cmap = 'Blues', annot = True)
plt.title('After PCA')
plt.show()
```
