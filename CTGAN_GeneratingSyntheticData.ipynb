```python
from ctgan import CTGANSynthesizer
from ctgan import load_demo
import pandas as pd

#data = load_demo()

# Names of the columns that are discrete
discrete_columns = [
    'BolgeCografya',
    'BolgeTipi',
    'AraziBoyutu',
    'AracYogunluk',
    'AracOrtMobility',
    'KameraliSistem',
    'DataRateDurum',
    'ReliabilityDurum',
    'LatencyDurum',
    'SensorYogunluk',
    'SensorVeriIletim',
    #'Standart'
]

data = pd.read_csv("MyDrive/NextGCT-IRG/dataFeatureSet.csv", header = None)

data.columns = discrete_columns

labels = pd.read_csv("MyDrive/NextGCT-IRG/dataCLassLabels.csv", header = None)
labels.columns = ['Standart']

dataset = pd.concat([data, labels], axis = 1)

ctgan = CTGANSynthesizer(epochs=10)
ctgan.fit(dataset, discrete_columns)

# Synthetic copy
samples = ctgan.sample(1000)
```
